On commence par importer les librairies nécessaires  : 

```python
import random
import statistics as stat
import math
import numpy as np
import matplotlib.pyplot as plt
```

Ensuite on trace nos graphes : 

La loi géométrique : 

```python

def geometrical(p,k):
    return p*(1-p)**(k-1)
#print(geometrical(0.5,random.randint(3,10))*100 + 5)
L = []
for i in range(100):
    L.append(geometrical(0.5,random.randint(3,5))*100 + 5)
#print(L)
#print(math.sqrt(stat.variance(L)))

geomRange = np.arange(len(L))
plt.plot(geomRange, L)
plt.show()
```

La loi binomiale (occurence des sorties)

```python

K = []
for i in range(100):
    maxRange = 25
    p = random.random()
    k = random.random()*maxRange
    binom = np.random.binomial(k,p)
    #if(binom != 0 ):
    K.append(binom)
    
print(K)
J = np.arange(18)
_ = plt.hist(K, J)
plt.show()
```
